```python
#!usr/bin/env python3
import pandas as pd
from pathlib import Path
import praw
from tqdm.notebook import tqdm
import pandas as pd
import yfinance as yf
import re
```

```python
reddit_df = pd.read_csv('10y_reddit_data.csv.zst', encoding='utf-8')
len(reddit_df)
```

```python
stock_df = pd.read_csv('10y_stock_data.csv.zst', header=[0, 1], index_col=0)
len(stock_df)
```

```python
# Create timestamp indexes, normalize, and sort
reddit_df['timestamp'] = (
    pd.to_datetime(reddit_df.created_utc, unit='s', utc=True)
    .dt.normalize()
    )
reddit_df = reddit_df.set_index('timestamp').sort_index()

stock_df.index = pd.to_datetime(stock_df.index).normalize().tz_localize('UTC')
```

```python
# Align data to first Monday and remove null row
stock_df = stock_df.drop(stock_df.index.min())

min_date = min(reddit_df.index.min(), stock_df.index.min())
max_date = max(reddit_df.index.max(), stock_df.index.max())
```
